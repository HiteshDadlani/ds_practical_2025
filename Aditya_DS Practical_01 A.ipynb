# Data Normalization using Z Score                                                                             

```python
#Z-score is a normalization technique which tells the distribution of the data in the dataset means it calcuates the mean
#and standard deviation of the numeric columns and then calculates the z score.Basically it tells that how far are values
#of that particular colum(where we have applied Z score normalization)from the average of that column
#Z score centers the values near to zero and it helps to detects the outliers in the dataset
#The formula to calculate z score is  
#                            z score=(xâˆ’Î¼)/Ïƒ
#where x is the value in that particular column and Î¼ is the mean of that particular column and Ïƒ is standard
#deviation of that particular column
                                
from sklearn.preprocessing import StandardScaler#standard scaler is a tool from sci-kit learn library which helps in 
#normalization of the data
numeric_cols = ['Salary', 'Bonus %']#we can only do normalization of numeric columns so in our dataset there are only 
#two numeric columns which are Salary and Bonus %
scaler = StandardScaler()#creating the object of Standard Scaler
# Apply Z-score normalization
df[numeric_cols] = scaler.fit_transform(df[numeric_cols])#fit removes the mean and Standard deviation of Salary and Bonus %
#column and transform calculates the z score 
# Display updated value
df
#Like for example in Salary column you will see that now first value is 0.201956 now lets understand how these value was 
#removed now we know that mean(average)of the salary column is 90662.181(from df.describe()command which is written above)
#and in dataset before applying the z score normalization the first value of Salary column was 97308 which is the value of x
#here and we also get to know the standard deviation of Salary column(from df.describe() command which is written above)
#standard deviation of Salary column is 32923.693342 so we calculate z score as (97308-90662.181)/32923.693342 which comes
#out to to 0.201956
#so same thing is done for each value of the Salary column and the Bonus % column
#If the z score value is positive then the original value(means value before applying z score)in dataset was above average 
#of the column and if the z score value is negative then the original value was below average of the column
```

```python
df.describe()
```

```python
import matplotlib.pyplot as plt#used for plotting basic graphs
import seaborn as sns#used to make plot attractive
# Create a box plot for normalized Salary
plt.figure(figsize=(6, 4))#this sets the size of the plot.Here we have taken size as 6 inches wide and 4 inches tall
sns.boxplot(y=df['Salary'])#We are plotting box plot for Salary column and it will be vertical as we have written y=
#if I had written x= then it will be horizontal
plt.title('Box Plot of Salary')#This is the title of the plot
plt.ylabel('Z-score')#We are adding label in y axis named "Z-score".
plt.grid(True)#to show grids in the graph
plt.show()#to display the plot
#Now in this box plot there are now outliers so you will not see any outliers
#Now you will wonder that we have some z score values above 1.0 or -1.0 but in box plot the blue part is between 0.5 ans 1 
#only the reason being that box plot plots the IQR Range means the 75% and 25%
#you see in the above df.describe()command that 25%(Q1)after Z score is -8.523714e-01 means-8.523714x10^-1 which actually 
#means its actually -0.8523714 and similarly for Q3(75%)which is 8.532493e-01 means 0.8523493 so box plot plots these values
#the lower bound(Q1)is -0.8523714 so blue part of the box plot ends between -0.5 and -1 as Q1 value(lower bound) value 
#is -0.852371 similarly for Q3

#we have written z-score in y axis because we are plotting the IQR range of z-score of Salary column
#There are no outliers in our box plot because they are no extreme or very different(outliers) z score values in our 
#dataset for Salary column

#We will see outliers only if there were z score values in Salary column above 3.41 or-3.409 
#I have shown how below how 3.41 or -3.409 have come
#Q1 â‰ˆ -0.852(from df.describe()command above)
#Q3 â‰ˆ +0.853(from df.describe()command above)
#â†’IQR = Q3 âˆ’ Q1 â‰ˆ 1.705
#Lower bound =-0.852âˆ’1.5Ã—1.705=-3.409(Lower Bound=Q1-1.5*IQR this is the formula)
#Upper bound=+0.853+1.5Ã—1.705=+3.419(Upper Bound=Q3-1.5*IQR this is the formula)
```

# Converting Categorical Values into Numerical Values

```python
df["Gender"]=df["Gender"].replace({"Female":0,"Male":1})#converting categorical data into numerical by assigning zero 
#to female and 1 to male
df.head()
```

```python
#Now as you can see that Gender Column has values 1.0 and 0.0 means floating values and data type of of Gender is 
#float and you want to convert it into integer so you will try to convert data type 
#of Gender but if you write the command df["Gender"]=df["Gender"].astype('int') you will
#get an error try it
#the reason for this is that we have null values in Gender column and we cannot convert null values into int data type
#If we wanted to convert to data type of Gender from float to any other data type instead of int there will be no error
#Because only int cant work with null values
#So drop the missing values from the dataset/from the Gender column or fill it we will drop null values from entire dataset
#using dropna() method
df=df.dropna()#dropna() method will remove the rows which contain atleast one null value
#Now write 
df.info()#we will see that all null values rows are dropped from the dataset and we are left with 764 rows now 
#instead of 1000
```

```python
df["Gender"]=df["Gender"].astype('int')
df.info()#Now you see that no error is there and the data type is also converted
```
