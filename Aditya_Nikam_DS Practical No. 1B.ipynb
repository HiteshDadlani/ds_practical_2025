# Data Normalization with MinMax Normalization

```python
#Min Max Normalization technique is a type of Normalization technique in which we scale all the values of the column 
#specified by us between 0 and 1 
#Its formula is                      new value=(X-min)/(Max-Min)

#where X=value in the column,min is the minimum value of that particular column and max is the maximum value of that 
#particular column

#We use it so that machine learning model learn quickly and do not give priority to only larger values and give less 
#importance to smaller values all values should be treated the same

numeric_col= ["WT", "AI", "DSBDA", "Average"]#We do normalization only of numeric column and these are the numeric 
#columns of our dataset we did not took Roll No because there is no use of Roll No column here

#Applying Min-Max Normalization
for col in numeric_col:#the loop will travel to each column mentioned in numeric_col one by one
    df[col]=(df[col]-df[col].min())/(df[col].max()-df[col].min())
    
#Now I will tell you the working of this so we have four numeric columns so for loop will visit every numeric column 
#mentioned in numeric_col one by one so it will first visit the WT column so it will pick the first value of WT column
#the picked value will be our x value like if we see WT column in our dataset the first value is 72 so that will be our 
#x value then minimum value of WT column is found out which is 18(from df.describe()command written above) and then
#we will find the maximum value of WT column which is 97(from df.describe()command written above) so now we will put these
#values in Min Max Normalization formula to get the answer new value=(72-18)/(97-18) which comes out to be 0.683544
#same thing will happen for each value of WT column and then for other column values as well

df#Now you will see the values scaled between 0 and 1
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(6, 4))
sns.boxplot(y=df['WT'])
plt.title("Box Plot of WT (After Min-Max Normalization)")
plt.ylabel("Normalized Marks (0 to 1)")
plt.grid(True)
plt.show()
#Now in this box plot there is an outliers in bottom side of the plot denoted by diamond shape 
#Now you will wonder that we have some new values above near to 0.8 or -0.8 but in box plot the blue part is between 0.6 
#and 0.8 only the reason being that box plot plots the IQR Range means the 75% and 25%
#you see in the above df.describe()command that 25%(Q1)after Min Max Normalization for WT is 0.43
#and similarly for Q3(75%)which is 0.69 so box plot plots these values
#the lower bound(Q1)is 0.43 so blue part of the box plot ends between between 0.6 and 0.8 as Q1 value(lower bound) value 
#is 0.43 and similarly for Q3

#we have written Normalized Marks in y axis because we are plotting the IQR range of Normalized Marks of WT column
#There are is an outliers in our box plot in bottom side of the plot denoted by diamond shape 

#We will see outliers only if there were z score values in Salary column above 1.03 or below 0.07
#I have shown how below how 1.03 or 0.07 have come
#Q1 ≈ 0.43(from df.describe()command above)
#Q3 ≈ 0.67(from df.describe()command above)
#→IQR = Q3 − Q1 =0.24
#Lower bound =0.43−1.5×0.24=0.07(Lower Bound=Q1-1.5*IQR this is the formula)
#Upper bound=0.67+1.5×0.24=+3.419(Upper Bound=Q3-1.5*IQR this is the formula)
```
