```python
df['age'] = df['age'].fillna(np.mean(df['age']))
df#We are replacing the missing values of age column with the mean value of age column
```

```python
df.isnull().sum()#This displays the number of null values in each column
```

```python
plt.figure(figsize=(10, 6))  # Set the figure size

# Create the box plot
sns.boxplot(x='sex', y='age', hue='survived', data=df)

# Add title and axis labels
plt.title('Age Distribution by Gender and Survival')
plt.xlabel('Gender')
plt.ylabel('Age')

# Show the plot
plt.show()

#The black line of each box plot represents the median age of that respective box plot.Means for example from the plot we 
#can see that median age of the male who survived was 30 or near to it and the male who died their median age was also 30 or
#near to it

#The observations for this plot are that we can see that:-

#1)Many older men did not survived as there are many outliers for survived box plot of men and the ages are also more

#2)The middle black line(median age) inside all boxes is around 28-30 years for both males and females, 
#regardless of survival

#3)There were some very young children also who did not survived(see the outliers of blue box for men who did not survived)

#4)Males who survived had a wider range of ages(see the box plot of survived for men(orange box plot) there you see that age
#ranges for men who survived starts from near 20 and continues till between 35 and 40 but more female there is not as much
#variation as men as female who survived box plot age starts between 20 and 30 and goes till between 30 and 40.
```
