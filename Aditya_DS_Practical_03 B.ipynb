```python
sns.histplot(df["SALARY"],bins=[1000,5000,10000,15000,20000,25000],kde=True,color='skyblue')#we are using histogram plot 
#which shows the distibution of data.I have used bins=[1000,5000,10000,15000,20000,25000] because we know that minimum 
#value in salary column is 1000 and maximum value is 25000 so if i used bins=10 instead of bins=[1000,5000,10000,15000,20000,
#25000] the data was not organized in x axis the number on x matches were not on the edges of the bar so it was difficult
#to study the graph so I manually give ranges from 1000 to 25000 with interval of 5000 so that we can easily study the graph

#bins=10 will divide salary into ten equal parts here we we have had minimum value of Salary as 2100 and 24000
#(from df["Salary"].max() and df["Salary"].min() command) so it will divide this range into 10 equal parts but in x axis 
#the numbers were not properly organized so we assigned bins as per our choice

#kde means kernel density estimation(kde) which is used to draw a line graph on histogram.It can be used to see the 
#skewness of our data we have used kde=True to plot the line graph on the histogram if we write kde=False then the line 
#graph will be removed.

#color='skyblue' means we have given color to bar graphs of histograms as skyblue

plt.title("Salary Distribution")#title of the plot
plt.xlabel("Salary")#name on x axis
plt.ylabel("Frequency")#name on y axis
plt.show()#to display the plot

#From the graph we can see that the data is positively skewed as tail is more towards the right
```
